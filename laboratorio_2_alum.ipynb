El dataset de viviendas de Boston contiene varias columnas que se utilizan para describir diversos aspectos de las viviendas residenciales de Boston. A continuación, se incluye una descripción de cada columna del dataset:

**1. CRIM:** Tasa de criminalidad per cápita por localidad. Indica el nivel de criminalidad en la zona.

**2. ZN:** Proporción de terrenos residenciales destinados a lotes de más de 25.000 pies cuadrados. Esta característica refleja la densidad residencial del área.

**3. INDUS:** Proporción de acres de terrenos destinados a negocios no minoristas por ciudad. Este es un indicador del uso comercial de terrenos fuera de las áreas residenciales.

**4. CHAS:** variable ficticia de Charles River (1 si el terreno limita con el río; 0 en caso contrario). Esto indica si la propiedad está cerca del río Charles, lo que puede agregar valor estético al vecindario.

**5. NOX:** Concentración de óxidos nítricos (partes por 10 millones). Representa el nivel de contaminantes industriales en la zona.

**6. RM:** Número medio de habitaciones por vivienda. Más habitaciones suelen indicar alojamiento más espacioso.

**7. AGE:** Proporción de unidades ocupadas por sus propietarios construidas antes de 1940. Las estructuras más antiguas pueden carecer de comodidades más nuevas o podrían considerarse más prestigiosas dependiendo de la arquitectura y la condición.

**8. DIS:** Distancias ponderadas a cinco centros de empleo de Boston. Esta característica mide la accesibilidad a los lugares de trabajo, lo que puede influir en los precios de la vivienda.

**9. RAD:** Índice de accesibilidad a carreteras radiales. Los valores más altos indican un acceso más fácil a las vías principales.

**10. TAX:** Tasa de impuesto a la propiedad sobre el valor total de la misma por cada $10,000. Refleja la tasa anual de impuesto a la propiedad.

**11. PTRATIO:** Ratio de alumnos por profesor por localidad. Los valores más bajos suelen indicar mejores instalaciones educativas.

**12. B:** Medida del porcentaje de población afroamericana

**13.LSTAT:** Porcentaje de población con bajo estatus socioeconómico.

**14. PRICE:**  Precio de las casas en miles de dólares.

### 1. Valores Faltantes por columna

```python
# Porcentaje de valores faltantes por cada columna

def revisar_valores_faltantes(df):
    # Calcular valores faltantes y porcentaje
    data_faltante = df.isnull().sum()
    data_faltante_porcentaje = (data_faltante / len(df)) * 100

    # Crear tabla
    data_faltante_tabla = pd.DataFrame({
        'Columna': data_faltante.index,
        'Valores Faltantes': data_faltante.values,
        'Porcentaje Faltantes (%)': data_faltante_porcentaje.values
    })

    # Ordenar por porcentaje de valores faltantes en forma descendente
    data_faltante_tabla = data_faltante_tabla.sort_values(by='Porcentaje Faltantes (%)', ascending=False).reset_index(drop=True)

    return data_faltante_tabla

print("Tabla de Valores Faltantes por columna")
revisar_valores_faltantes(braziltourism)
```

```python
print("Gráfico de Valores Faltantes por columna")
plt.figure(figsize=(8, 4))
# Se define el gráfico como barplot
barplot = sns.barplot(data=revisar_valores_faltantes(braziltourism), x='Columna', y='Valores Faltantes')
# Se agregan los valores para cada columna al barplot
for container in barplot.containers:
    barplot.bar_label(container)
plt.xticks(rotation=90)
#plt.xlabel('Columnas')
#plt.ylabel('Valores Faltantes por columna')
plt.title('Valores Faltantes por Columna')
plt.show()
```

De análisis de las variables se concluye:

* La variable numérica **Age** tiene una distribución relativamente simétrica con una media de 34.27 y mediana de 32.0 cercanas. Considerando esto, y que solamente se tiene un valor faltante, se realizará imputación con el valor de la **media**
* La variable numérica **Travel_cost** muestra una distribución sesgada hacia la derecha con presencia de outliers. La media de 115.13 y mediana de 48.00 no se encuentran cercanas. Consideranto esto, y que solamente se tiene 3 valores faltantes (0.7%), se realizará imputación con el valor de la **mediana** para manejar mejor el sesgo y outliers.
* La variable numérica **Logged_income** muestra una distribución más simétrica con los valores de media de 6.63 y mediana de 8.84 cercanos. Se observan algunos outliers, por lo que se realizará imputación con el valor de la **moda**

```python
from sklearn.impute import SimpleImputer

# Crear imputadores para la media, mediana y moda
imputador_media = SimpleImputer(strategy='mean')
imputador_mediana = SimpleImputer(strategy='median')
imputador_moda = SimpleImputer(strategy='most_frequent')

# Realizar imputación de variables categóricas
braziltourism['Sex'] = imputador_moda.fit_transform(braziltourism[['Sex']]).ravel()
braziltourism['Access_road'] = imputador_moda.fit_transform(braziltourism[['Access_road']]).ravel()
braziltourism['Active'] = imputador_moda.fit_transform(braziltourism[['Active']]).ravel()
braziltourism['Passive'] = imputador_moda.fit_transform(braziltourism[['Passive']]).ravel()
braziltourism['Trips'] = imputador_moda.fit_transform(braziltourism[['Trips']]).ravel()

# Realizar imputación de variables numéricas
braziltourism['Age'] = imputador_media.fit_transform(braziltourism[['Age']]).ravel()
braziltourism['Travel_cost'] = imputador_mediana.fit_transform(braziltourism[['Travel_cost']]).ravel()
braziltourism['Logged_income'] = imputador_media.fit_transform(braziltourism[['Logged_income']]).ravel()

# Validar que no existan valores faltantes para las variables imputadas
print("Tabla de Valores Faltantes por columna")
revisar_valores_faltantes(braziltourism)
```

3.d Imputación con Regresión Lineal

```python
import pandas as pd
from sklearn.experimental import enable_iterative_imputer
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.impute import IterativeImputer

features = ['Age', 'Sex', 'Travel_cost', 'Access_road', 'Active',	'Passive',	'Logged_income', 'Trips']
df_regression = braziltourism[features + ['Income']].dropna()

X = df_regression[features]
y = df_regression['Income']

regressor = LinearRegression()
regressor.fit(X, y)

# Entrenar modelo de regresión
regressor = LinearRegression()
regressor.fit(X, y)

# Predecir valores faltantes
def predecir_income_reg(row):
    if pd.isnull(row['Income']):
        row_data = pd.DataFrame([[row['Age'], row['Sex'], row['Travel_cost'], row['Access_road'], row['Active'], row['Passive'], row['Logged_income'], row['Trips']]], columns=features)
        return regressor.predict(row_data)[0]
    return row['Income']

braziltourism['Income_regression'] = braziltourism.apply(predecir_income_reg, axis=1)
```

3.e Imputación con KNN

```python
# Entrenar modelo KNN
imputador_knn = KNeighborsRegressor(n_neighbors=5)
imputador_knn.fit(X, y)

# Predecir valores faltantes
def predecir_income_knn(row):
    if pd.isnull(row['Income']):
        row_data = pd.DataFrame([[row['Age'], row['Sex'], row['Travel_cost'], row['Access_road'], row['Active'], row['Passive'], row['Logged_income'], row['Trips']]], columns=features)
        return imputador_knn.predict(row_data)[0]
    return row['Income']

braziltourism['Income_knn'] = braziltourism.apply(predecir_income_knn, axis=1)
```

3.f Imputación con MICE

```python
# Entrenar modelo MICE
imputador_mice = IterativeImputer(max_iter=10, random_state=42)
imputados_mice = imputador_mice.fit_transform(braziltourism[features + ['Income']])

# Crear la columna imputada con MICE
braziltourism['Income_mice'] = imputados_mice[:, -1]
```

### 4.2 Métodos estadísticos univariados, limitaciones y transformaciones (4 puntos)

Para distribuciones con colas largas, los métodos tradicionales de detección de atípicos como la desviación estándar o el rango intercuartil (IQR) pueden no ser adecuados, ya que pueden clasificar erróneamente valores extremos como outliers cuando en realidad forman parte de la distribución natural.

### 4.2.1 (1 puntos)
¿Cómo podría ajustar cada uno de **estos dos métodos descritos anteriormente** para hacerlos más robustos frente a estos casos de distribuciones asimétricas de largas colas?

Los métodos de bloxplot y z-score se pueden modificar de la siguiente forma para que sean más robustos a distribuciones asimetricas con colas largas:

* **Boxplot modificado**: modifica los límites de los valores atípicos introduciendo el MedCouple para escalarlos en función de la asimetría:

  * Límite superior: Q3 + (1,5 × e ^ (3 × MC) × RIQ)
  * Límite inferior: Q1 − (1,5 × e ^ (3 × MC) × RIQ)

  *MedCouple (MC)* es una medida robusta y no paramétrica de asimetría, diseñada específicamente para su uso con datos asimétricos, resistente a los valores atípicos, lo que la hace ideal para distribuciones con colas pesadas o irregularidades.

* **Z-score modificado**: modifica el Z-score con métricas más robustas para asimetria:

  * Calcula la distancia de las observaciones desde la mediana (en vez de la media), que es más estable a la influencia de valores atípicos.
  * Utiliza la desviación absoluta media (MAD), que es la mediana de las desviaciones absolutas de la mediana, lo que la hace más resistente a la influencia de datos extremos.
    * mz = 0.6745 * (xi - mediana) / MAD
    * MAD = mediana(xi - mediana)

b.1) Criterio del boxplot

```python
# Grafico de Cajas para visualizar existencia de atípicos
plt.figure(figsize = (4,8))
sns.boxplot(y = salaries_2024_ft['TotalPayBenefits'])
plt.title("Grafico de Boxplot de TotalPayBenefits")
```

```python
# Calcular IQR, Q1, Q2, umbral, limite superior y limite inferior

def atipicos_boxplot(df , columna):
  Q1 = df[columna].quantile(0.25)
  Q3 = df[columna].quantile(0.75)
  IQR = Q3 - Q1
  # Calcular el umbral para los valores atípicos
  umbral_atipico = IQR * 1.5
  # calcular el valor del límite inferior y superior
  limite_inferior = Q1 - umbral_atipico
  limite_superior = Q3 + umbral_atipico
  # Calcular el atipicos debajo del límite inferior y por encima del límite superior
  df_inferior = df[df[columna] > limite_superior]
  df_superior = df[df[columna] < limite_inferior]
  return IQR, df_inferior, df_superior, limite_inferior, limite_superior


iqr, df_inf, df_sup, lim_inf, lim_sup = atipicos_boxplot(salaries_2024_ft, 'TotalPayBenefits')
print('El rango intercuartil es: ' + str(iqr))
print('El limite inferior es: ' + str(lim_inf))
print('El limite superior es: ' + str(lim_sup))
print('El número total de valores atípicos es:', df_inf.shape[0] + df_sup.shape[0])
#
```

```python
# Histograma con rangos de atipicos
plt.figure(figsize = (10,6))
sns.histplot(salaries_2024_ft['TotalPayBenefits'], kde=False)
plt.axvspan(xmin = lim_inf,xmax= salaries_2024_ft['TotalPayBenefits'].min(),alpha=0.2, color='red')
plt.axvspan(xmin = lim_sup,xmax= salaries_2024_ft['TotalPayBenefits'].max(),alpha=0.2, color='red')
plt.title('Histograma de TotalPayBenefits con rangos de atipicos boxplot')
plt.xlabel('TotalPayBenefits')
plt.show()
```

b.2) Criterio del Z-score

```python
# Calcular atipicos y zscore

def atipicos_zscore(data,columna):
    #global outliers,zscore
    atipicos = []
    zscore = []
    umbral_zscore = 3
    media = np.mean(data[columna])
    std = np.std(data[columna])
    for i in data[columna]:
        z_score= (i - media)/std
        zscore.append(z_score)
        if np.abs(z_score) > umbral_zscore:
            atipicos.append(i)
    return atipicos, zscore

atipicos, zscore = atipicos_zscore(salaries_2024_ft, 'TotalPayBenefits')
print("El número total de atípicos con Z-score es: ",len(atipicos))
```

```python
# Histograma con rangos de atipicos zscore
plt.figure(figsize = (10,6))
sns.histplot(zscore)
plt.axvspan(xmin = 3 ,xmax= max(zscore),alpha=0.2, color='red')
plt.title('Histograma de TotalPayBenefits con rangos de atipicos zscore')
plt.xlabel('TotalPayBenefits')
plt.show()
```

```python
import statsmodels.api as sm

def atipicos_boxplot_modif(df , columna):
  # Calcular rango intercuartil
  Q1 = df[columna].quantile(0.25)
  Q3 = df[columna].quantile(0.75)
  IQR = Q3 - Q1
  # Calcular el umbral para los valores atípicos
  #umbral_atipico = IQR * 1.5
  # Calcular el MC (Medcouple)
  MC = sm.stats.stattools.medcouple(df[columna])
  # calcular el valor del límite inferior y superior
  if MC > 0:
      limite_inferior = Q1 - 1.5 * np.exp(-4 * MC) * IQR
      limite_superior = Q3 + 1.5 * np.exp(3 * MC) * IQR
  else:
      limite_inferior = Q1 - 1.5 * np.exp(-3 * MC) * IQR
      limite_superior = Q3 + 1.5 * np.exp(4 * MC) * IQR
  # Calcular los atipicos debajo del límite inferior y por encima del límite superior
  df_inferior = df[df[columna] > limite_superior]
  df_superior = df[df[columna] < limite_inferior]
  return IQR, df_inferior, df_superior, limite_inferior, limite_superior


iqr, df_inf, df_sup, lim_inf, lim_sup = atipicos_boxplot_modif(salaries_2024_ft, 'TotalPayBenefits')
print('El rango intercuartil es: ' + str(iqr))
print('El limite inferior es: ' + str(lim_inf))
print('El limite superior es: ' + str(lim_sup))
print('El número total de valores atípicos es:', df_inf.shape[0] + df_sup.shape[0])
#
```
